   - `Conv2D(128, (3, 3), strides=(1, 1), padding='valid')`: This is the fifth Convolutional layer with 128 filters of size (3, 3), using 'valid' padding, and a stride of (1, 1). The activation function used in this layer is 'relu'.

3. MaxPooling Layers:
   - `MaxPooling2D((2, 2))`: This is the first MaxPooling layer with a pool size of (2, 2). It performs downsampling by taking the maximum value in each 2x2 region.
